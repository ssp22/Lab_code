```python
import numpy as np
import scipy as sp
import matplotlib 
from numpy import genfromtxt
from matplotlib import pyplot as plt
```

```python
my_data = genfromtxt('sample4Phadke.txt', delimiter='	')
sample_1 = my_data[1:9,:]
sample_2 = my_data[9:17,:]
sample_3 = my_data[17:25,:]
sample_4 = my_data[25:33,:]

sampleZ_1 = sample_1[:,2]
sampleZ_2 = sample_2[:,2]
sampleZ_3 = sample_3[:,2]
sampleZ_4 = sample_4[:,2]
SampleArray = [sampleZ_1,sampleZ_2,sampleZ_3,sampleZ_4]
plt.xlabel('Samples')
plt.ylabel('Particle Size in nm')
plt.boxplot(SampleArray,showmeans=True)
zMEANS = [np.mean(sampleZ_1),np.mean(sampleZ_2),np.mean(sampleZ_3),np.mean(sampleZ_4)]
print(zMEANS)
plt.savefig('vis_boxplot')
plt.figure(2)
plt.xlabel('Samples')
plt.ylabel('Mean Particle Size in nm')
plt.plot([1,2,3,4],zMEANS)
plt.plot([1,2,3,4],[18,18,2,2])
plt.plot([1,2,3,4],[100,1,1,100])
```

```python
sampleP_1 = sample_1[:,3]
sampleP_2 = sample_2[:,3]
sampleP_3 = sample_3[:,3]
sampleP_4 = sample_4[:,3]
SampleArray = [sampleP_1,sampleP_2,sampleP_3,sampleP_4]
plt.xlabel('Samples')
plt.ylabel('Polydispersity index')
plt.boxplot(SampleArray,showmeans=True)

plt.savefig('vis_PDI')
```

```python
sampleZ_1 = sample_1[:,4]
sampleZ_2 = sample_2[:,4]
sampleZ_3 = sample_3[:,4]
sampleZ_4 = sample_4[:,4]
SampleArray = [sampleZ_1,sampleZ_2,sampleZ_3,sampleZ_4]
plt.xlabel('Samples')
plt.ylabel('Number Mean Particle Size in m,')
plt.boxplot(SampleArray,showmeans=True)

plt.savefig('vis_NM')
```
